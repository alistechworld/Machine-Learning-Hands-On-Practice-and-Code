```python
def rename(name):
  if name == 'Royal Challengers Bangalore':
    return "RCB"
  else:
    return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
# Convertors
pd.read_csv('IPL Matches 2008-2020 (1).csv', converters={'team1': rename})
```
